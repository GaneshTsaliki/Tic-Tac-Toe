#### CASE 1 : PLAYER 1 PLAYS AGAINST PLAYER 2 GIVING INPUT MANUALLY

```python
# player 1 plays X and player 2 playes O
# Size of TIC-TAC-TOE boxes
N = 3

# defining the field
field = [['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']]

row1 = field[0]
row2 = field[1]
row3 = field[2]

col1 = []
col2 = []
col3 = []
diag1 = []
diag2 = []

gameover = False

def colUpdate():
# Update the column every round
    col1.clear()
    col2.clear()
    col3.clear()
    for row in field:
        col1.append(row[0])
    for row in field:
        col2.append(row[1])
    for row in field:
        col3.append(row[2])
        
def diagUpdate():
# Update the Diagonals every round
    diag1.clear()
    diag2.clear()
    for i in range(N):
        diag1.append(field[i][i])
    for j in range(N):
        diag2.append(field[j][N-1-j])

def getUpdatePlayer1():
    p1row = int(input("\nPlayer 1 Row: "))
    p1col = int(input("Player 1 Column: "))
    if (field[p1row-1][p1col-1]) == '-':
        field[p1row-1][p1col-1] = 'X'
        colUpdate()
        diagUpdate()
    else:
        print("\nNO CHEATING!\n")
        getUpdatePlayer1()
            
    print(field[0])
    print(field[1])
    print(field[2])
    return col1
    return col2
    return col3
    return diag1
    return diag2

def getUpdatePlayer2():
    p2row = int(input("\nPlayer 2 Row: "))
    p2col = int(input("Player 2 Column: "))
    if (field[p2row-1][p2col-1]) == '-':
        field[p2row-1][p2col-1] = 'O'
        colUpdate()
        diagUpdate()
    else:
        print("\nNO CHEATING!\n")
        getUpdatePlayer2()
            
    print(field[0])
    print(field[1])
    print(field[2])
    return col1
    return col2
    return col3
    return diag1
    return diag2

def checkWinner():
    x = ['X', 'X', 'X']
    o = ['O', 'O', 'O']
    global gameover
    
    if (row1 == x or row2 == x or row3 == x or col1 == x or col2 == x or col3 == x or diag1 == x or diag2 == x):
        print("PLAYER 1 WINS THIS ROUND!")
        gameover = True
    elif (row1 == o or row2 == o or row3 == o or col1 == o or col2 == o or col3 == o or diag1 == o or diag2 == o):
        print("PLAYER 2 WINS THIS ROUND!")
        gameover = True
    elif (('-' not in row1) and ('-' not in row2) and ('-' not in row3)):
        print("\nDRAW MATCH!\n")
        gameover = True
    return

def startGame():
    print("PLAYER 1 PLAYS AS 'X' and PLAYER 2 PLAYS AS 'O'")
    n = 4
    while (not gameover):
        getUpdatePlayer1()
        checkWinner()        
        if (gameover):
            break
        getUpdatePlayer2()
        checkWinner()
        if (gameover):
            break
```

```python
startGame()
```

#### CASE 2 : GIVEN A PREDEFINED GAME OF TIC TAC TOE, WE ANALYSE THE OUTCOME
